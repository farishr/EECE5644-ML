```python
import matplotlib.pyplot as plt

def calculate_bollinger_bands(data, window_size, num_std_dev):
    upper_band, lower_band = [], []

    for i in range(len(data)):
        if i < window_size:
            upper_band.append(None)
            lower_band.append(None)
        else:
            window_data = data[i - window_size + 1:i + 1]
            rolling_mean = sum(window_data) / window_size
            rolling_std = (sum([(x - rolling_mean) ** 2 for x in window_data]) / window_size) ** 0.5
            upper_band.append(rolling_mean + (num_std_dev * rolling_std))
            lower_band.append(rolling_mean - (num_std_dev * rolling_std))

    return upper_band, lower_band

# Sample data
dates = ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05']
closing_prices = [100, 105, 98, 102, 104]

# Set parameters
window_size = 3  # Adjust according to your preference
num_std_dev = 2  # Adjust according to your preference

# Calculate Bollinger Bands
upper_band, lower_band = calculate_bollinger_bands(closing_prices, window_size, num_std_dev)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(dates, closing_prices, label='Closing Price')
plt.plot(dates[window_size - 1:], upper_band, label='Upper Bollinger Band', linestyle='--', color='red')
plt.plot(dates[window_size - 1:], lower_band, label='Lower Bollinger Band', linestyle='--', color='green')
plt.fill_between(dates[window_size - 1:], upper_band, lower_band, color='lightgray', alpha=0.4, label='Bollinger Bands')
plt.title('Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
